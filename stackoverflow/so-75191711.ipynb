```python
import pandas as pd
import seaborn as sns
import seaborn.objects as so
from matplotlib.dates import DateFormatter

sns.set_theme()
```

```python
df = pd.read_csv("so-75191711.csv")
list(df["Unnamed: 0"].astype(float).index.values)
```

```python
df.count().Weight
```

```python
df2 = df.set_index(pd.date_range(start="2/4/2022", periods=df.count().Weight, name="Date"))
```

```python
(
    so.Plot(df2, x="Date", y="Weight")
        .add(so.Line())
        .scale(x=so.Temporal().label(formatter=DateFormatter('%m-%d')))
        .label(title="Daily Weight February 2022")
)
```
